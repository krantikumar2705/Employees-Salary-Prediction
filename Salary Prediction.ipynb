# label Encoding 

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
```

```python
df['Job_enc'] = lb.fit_transform(df['Job'])
```

```python
df['Gender_enc'] = lb.fit_transform(df['Gender'])
```

```python
df['Education_enc'] = lb.fit_transform(df['Education'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler`
```

```python
ss = StandardScaler()
```

```python
df['Age_scal'] = ss.fit_transform(df[['Age']])
```

```python
df['Experience_scal'] = ss.fit_transform(df[['Experience']])
```

```python
df.head()
```

Standard deviation is a measure of the dispersion or variability of a set of values from their mean, indicating how much the values deviate from the average.  and standarlisatoin ,mean the mean is 0 and standard deviation is 1.

```python
x = df[['Age_scal','Job_enc','Gender_enc','Education_enc','Experience_scal']]
y = df['Salary']
```

```python
# splitting the data in trainin and testing 
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state =  42)
```

```python
x_train.shape,y_train.shape #80% data for training
```

```python
x_test.shape,y_test.shape #20% data for testing validation
```

#### model development

```python
from sklearn.linear_model import LinearRegression
lr_model =LinearRegression() 
```

#### model training

```python
lr_model.fit(x_train,y_train)
```

#### model Prediction

```python
y_prediction_LR = lr_model.predict(x_test)
```

```python
y_prediction_LR # you are getting the 65 value be 65 record we used for the testing this is the true value
```

```python
df = pd.DataFrame({'y_actual': y_test,'y_predict':y_prediction_LR})
```

```python
df.head()
```

```python
# i want to calcultate differnce betn 
df['Error'] = df['y_actual']-df['y_predict'] # error shoub be absolute
df['abs_Error']  = abs(df['Error'])
```

```python
df.head()
```

```python
mean_abs_error = df['abs_Error'].mean() #manual calculte
```

```python
mean_abs_error
```

#### what is the accuracy of this model we are checking

```python
from sklearn.metrics import accuracy_score,r2_score
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
r2_score(y_test,y_prediction_LR) #89 % of accuracy
```

```python
round(mean_absolute_error(y_test,y_prediction_LR),2) # mean absolute error
```

## mean squared error

```python
mse  = round(mean_squared_error(y_test,y_prediction_LR),2)
mse
```
